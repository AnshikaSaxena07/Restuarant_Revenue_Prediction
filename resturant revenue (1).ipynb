```python
def encode_and_bind(df, feature):
    dummies = pd.get_dummies(df[[feature]], prefix='')
    return pd.concat([df, dummies], axis=1)
df = encode_and_bind(df, 'City')
df = encode_and_bind(df, 'City Group')
df = encode_and_bind(df, 'Type')
df.head(3)
```

<b>Model Building</b>

```python
df.head()
```

<b> Features and Target </b>

```python
x= df.drop(['Id','revenue', 'Open Date', 'City', 'City Group', 'Type','_Ankara','_Trabzon','_Tokat','_İstanbul','_Uşak','_İzmir','_Şanlıurfa','_Muğla','_Osmaniye','_Sakarya','_Samsun','_Tekirdağ','_Kayseri','_Kocaeli','_Konya','_Kütahya','_Kırklareli','_Eskişehir','_Gaziantep','_Isparta','_Karabük','_Kastamonu','_Bursa','_Denizli','_Diyarbakır','_Edirne','_Elazığ','_Amasya','_Antalya','_Aydın','_Balıkesir','_Bolu','City_cat','City Group_cat','Type_cat','_Adana','_Afyonkarahisar'], axis=1)
#X = df['_Big Cities','_Other','_DT','_IL','_FC'].values
y = df['revenue'].values
x.describe()
```

### <b>Spliting of Data</b>

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=1/3, random_state=202)
```

## <b>1. Linear Regression</b>

#### <b>Creating And Traing the Model</b>

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(xtrain, ytrain)

model.coef_
```

#### <b>Evaluation</b>

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
pred=model.score(xtrain,ytrain)
ypred =model.predict(xtest)
print(pred*100)
```

<b>Mean Absolute Error</b>

```python
r= mean_absolute_error(ytest, ypred) 
print("MAE: {0}".format(r))
```

<b>Mean Squared Error</b>

```python
rmse = mean_squared_error(ytest, ypred, squared=False)
print("RMSE: {0}".format(rmse))
```

## <b>2. Decision Tree Regression</b>

```python
from sklearn.tree import DecisionTreeRegressor 
```

```python
reg= DecisionTreeRegressor(max_depth=10)
reg.fit(xtrain, ytrain)
```

```python
pred = reg.score(xtrain,ytrain)
y_pred = reg.predict(xtest)
pred*100
```

<b>Mean Absolute Error</b>

```python
r1= mean_absolute_error(ytest, ypred) 
print("MAE: {0}".format(r1))
```

<b>Mean Squared Error</b>
